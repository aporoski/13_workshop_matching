### сокращение выборки

    Введем функцию reduce_df, которая будет уменьшать выборку base и и выбирать из train только те, ответы на которые есть в base. Позволит нам быстрее проводить расчеты.

```python
def reduce_df(df_base,df_train, frac_k, RANDOM_STATE):
    df_base_rs=df_base.sample(frac=frac_k, random_state=RANDOM_STATE)
    # применяю sample, но нет полной уверенности, что в итоге у нас будет репрезентативная выборка
    df_train_rs=df_train.query('Target in @df_base_rs.index') # выберем из train только те, target для которых есть в base
    print(f'size of df_base_sample{df_base_rs.shape}, size of df_train_sample {df_train_rs.shape}')
    return(df_base_rs,df_train_rs)
```

```python
df_base_sample,df_train_sample=reduce_df(df_base,df_train, 0.1, RANDOM_STATE)
```

### оптимизация кол-ва кластеров

    в данной подзадаче в качестве метода кластеризации используется MiniBatchKMeans, 
    так как использование обычного KMeans требует значительно больше времени
    для оценки будем использовать встроенные методы sklearn
    silhouette_score
    davies_bouldin_score

```python
def optimise_clusters(data,k_list):
    df_base=data
    db_scores=[]
    shl_scores=[]
    BATCH_SIZE=10000
    for i in tqdm(k_list):

    #     kmeans = KMeans(n_clusters=i, 
    #                     random_state=RANDOM_STATE,
    #                     init='k-means++',
    #                     ) 
    #     kmeans.fit(data)

        mkb=MiniBatchKMeans(n_clusters=i,
                           random_state=RANDOM_STATE,
                           init='k-means++',
    #                        n_init='auto', убрал, потому что падает с этим параметром,
                           batch_size=BATCH_SIZE) #компилятор советуе т поставить значение >=3072 чтобы избежать утечки памяти
        mbk_means_labels= mkb.fit_predict(df_base) 
        sh_score=round(silhouette_score(df_base, mbk_means_labels, random_state=RANDOM_STATE, sample_size=BATCH_SIZE),3)
        db_score=round(davies_bouldin_score(df_base, mbk_means_labels),3)
        db_scores.append(db_score)
        shl_scores.append(sh_score)

    #The value of the silhouette coefﬁcient is between [-1, 1]. 
    #A score of 1 denotes the best meaning that the data point i is very compact 
    #within the cluster to which it belongs and far away from the other clusters. The worst value is -1.
    #Values near 0 denote overlapping clusters.
    
    
    #getting scores together
    clusters=pd.DataFrame(data=np.array([k_list,db_scores,shl_scores]).T, columns=['clusters','db','sh'])
    
    #plot
    fig, ax1 = plt.subplots(figsize=(4,4))
    fig.suptitle("Зависимость оценок от кол-ва кластеров", fontsize=12)
    ax2 = ax1.twinx()
    plot1=ax1.plot(clusters.clusters,
                  clusters.db,
              label= 'davis-bolduine'
            )

    plot2=ax2.plot(clusters.clusters,
                  clusters.sh, color='red',
              label= 'silhoette'
                 )
    ax2.set_ylabel('silhouette')
    ax1.set_ylabel('davis-bolduine')

    obj=plot1+plot2
    label=[l.get_label() for l in obj]
    plt.legend(obj, label, loc=0)
```

```python
k_list=[50,75,90,100,110,125,150,175,200,250,500] # кол-во кластеров
optimise_clusters(df_base_sample,k_list)
```

### acc5

    Введем фукцию опеределения нашей метрики accuracy5

```python
def acc5(k,targets,idx,base_index): 
    acc=0
    if k==0: #faiss        
        for target, el in zip(targets.values.tolist(), idx.tolist()):
            acc += int(target in [base_index[r] for r in el])
        return 100 * acc / len(idx)
    elif k==1: #annoy
        for target, el in zip(targets.values.tolist(), idx):
            acc += int(target in [base_index[r] for r in el])
        return 100 * acc / len(idx)
```

```python
def scale(scaler_type,df_base, df_base_rs,df_train_rs, features):
    if scaler_type=='StandardScaler':
        Scaler=StandardScaler()
    elif scaler_type =='RobustScaler':
        Scaler=RobustScaler()
    elif scaler_type == 'MinMaxScaler':
        Scaler=MinMaxScaler()
    else:
        print('указан неправильный тип машстабирования')
    Scaler.fit(df_base[features]) # делаем fit на полной выборке
    df_base_rs_temp=df_base_rs.copy() 
    df_base_rs_temp[features]=Scaler.transform(df_base_rs_temp[features]) 
    
    df_train_rs_temp=df_train_rs.copy()
    
    target=df_train_rs_temp['Target'] # почему-то pandas выдает ошибку если не убрать target предварительно
    df_train_rs_temp=df_train_rs_temp.drop('Target',axis=1) # убираем
    df_train_rs_temp[features]=Scaler.transform(df_train_rs_temp[features])
    df_train_rs_temp['Target']=target # добавляем
    
    return df_base_rs_temp, df_train_rs_temp
```

#### FAISS

    Введем функцию, которая будет принимать на вход метрику для faiss, использование PCA или нет, n_probe, scaler, использование уменьшенной выборки, разные типы квантователей - не успел) которая будет выдавать нам accuracy, метрику, тип квантователя, тип масштабирования.  
    Затем протестируем её с разными параметрами, выберем наилучшие

```python
def get_acc5_faiss (df_base,df_train,n_clusters,n_probe, metric):    
#     exec_times=[]
    tic = time.time()
    dims = df_base.shape[1]
    n_cells = n_clusters 
    quantizer = faiss.IndexFlatL2(dims)    
    
    if metric=='INNER_PRODUCT':
        idx_l2=faiss.IndexIVFFlat(quantizer,dims,n_cells,faiss.METRIC_INNER_PRODUCT)
    elif metric=='METRIC_L2':
        idx_l2=faiss.IndexIVFFlat(quantizer,dims,n_cells,faiss.METRIC_L2)
    else:
        print('некорректная метрика')
    idx_l2.nprobe=n_probe #number of probes at query time
    idx_l2.train(np.ascontiguousarray(df_base.values[:50000, :]).astype('float32') )
    idx_l2.add(np.ascontiguousarray(df_base.values).astype('float32')) 
    base_index = {k: v for k, v in enumerate(df_base.index.to_list())}  
    try:
        targets = df_train['Target']
        df_train_temp=df_train.drop('Target', axis=1)
    except:
        pass
    r, idx = idx_l2.search(np.ascontiguousarray(df_train_temp.values).astype('float32'), 5) # проверимся на traine
    toc = time.time()
    print(f'faiss acc5: {acc5(0,targets,idx,base_index):.2f}, time: {toc - tic:.1f}, nprobe: {n_probe}, metric: {metric}')
    
#     %xdel 
```

#### PCA FAISS

        PCA применяется в данном случае в качестве альтернативы ручной уменьшения размеров выборки, для этого мы ранее ввели функцию  reduce_df(). В отличие от предыдущей фукнции faiss_test также изменяется структура. Используется IndexPreTransform. Ведем функцию для PCA от FAISS - PCA_faiss
    

```python
def PCA_faiss(dims,data,train):
#     exec_times=[]
    tic = time.time()
    
    coarse_quantizer = faiss.IndexFlatL2(dims) #Индекс, который хранит сокращенные до нужной размерности векторы
    
    for nlist in [5]: # было проведено тестировани при nlist=10, время расчета увеличивается почти в 2 раза, и 
        #точность уменьшается
        M, nbits = 10, 8
        # nlist - number of inverted lists
        # M -число подквантователей
        # Размерность вектора (d) должна быть кратна числу подквантователей (M)
        # nbits - 8 

        sub_index = faiss.IndexIVFPQ(coarse_quantizer,dims,nlist, M,nbits) # сокращаем индексы через IndexIVFPQ
        pca_matrix = faiss.PCAMatrix (len(features), dims, 0, True) # PCA-матрица
        index = faiss.IndexPreTransform (pca_matrix, sub_index) # преобразовываем индекс с учетом PCA-матриц
        index.train(data)  
        index.add(data)
        
        targets = train['Target']
        temp_df_train=train.drop('Target', axis = 1)
        
        r, idx = index.search(temp_df_train[features], 5)
        
        base_index = {k: v for k, v in enumerate(data.index.to_list())}
        toc = time.time()

        print(f'reduced dims:{k}, nlist:{nlist}, acc@5: {acc5(0,targets, idx, base_index):.2f}, exec_time: {toc - tic:.2f} seconds')
        
        
```

#### Annoy

    введем функцию для расчета соответствий по модели annoy с использованием метрики L2,
    оценки точности, и затрат времени. с учетом кол-ва кластеров, и кол-ва деревьев

```python
def get_acc5_annoy(df_base,df_train,n_clusters,annoy_n_trees):
    tic = time.time()
    annoy_distance_type = 'euclidean'
    k_similar = n_clusters
    annoy_search_n_nodes = -1
    dims = df_base.shape[1]
    
    base_index = {k: v for k, v in enumerate(df_base.index.to_list())}
    
    annoy_index = AnnoyIndex(dims, annoy_distance_type)
    annoy_index.set_seed(RANDOM_STATE)
    for i, v in zip(range(df_base.shape[0]), df_base.values):
         annoy_index.add_item(i, v)
            
    annoy_index.build(annoy_n_trees, n_jobs=-1)
     
    try:
        targets = df_train['Target']
        df_train_temp=df_train.drop('Target', axis=1)
    except:
        pass  
        
    idx = []
    for v in df_train_temp.values:
        idx.append(annoy_index.get_nns_by_vector(v, k_similar, search_k=annoy_search_n_nodes))
    toc = time.time()  
    print(f'annoy acc5: {acc5(1,targets,idx,base_index):.2f}, time: {toc - tic:.1f}, trees:{annoy_n_trees}')
#     %xdel
```
